```python
import requests

# Function to get weather data from OpenWeatherMap API using latitude and longitude
def get_weather_data(lat, lon):
    # Replace with your OpenWeatherMap API key
    api_key = '35879ab0c5fbcd67b3b649f3119fa468'
    base_url = 'http://api.openweathermap.org/data/2.5/weather'

    params = {
        'lat': lat,
        'lon': lon,
        'appid': api_key,
        'units': 'metric'  # Get temperature in Celsius
    }

    # Send a GET request to the OpenWeatherMap API
    response = requests.get(base_url, params=params)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        data = response.json()
        print(f"Weather for coordinates ({lat}, {lon}):")
        print(f"Temperature: {data['main']['temp']}°C")
        print(f"Weather: {data['weather'][0]['description']}")
        return data
    else:
        print("Location not found or invalid API key.")
        return None

# Main function to demonstrate the functionality
def main():
    # Part 1: Interacting with the OpenWeatherMap API using coordinates
    lat = input("Enter the latitude for weather data: ")
    lon = input("Enter the longitude for weather data: ")

    # Convert the input to float since latitude and longitude should be numbers
    try:
        lat = float(lat)
        lon = float(lon)
        get_weather_data(lat, lon)
    except ValueError:
        print("Invalid input! Latitude and longitude must be numbers.")

# Ensure the main function is executed when the script is run
if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup

# Simple web scraping function to fetch headlines from a news website
def scrape_news_headlines():
    url = 'https://www.bbc.com/news'  # Example news website
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'lxml')  # Use 'lxml' parser for faster parsing

        # Print out a part of the HTML to inspect its structure and understand where headlines are
        print(soup.prettify()[:1500])  # Print the first 1500 characters of HTML for inspection

        # Try searching for headlines using other tags or classes
        headlines = soup.find_all('a', class_='gs-c-promo-heading')  # Common class for links to articles

        if not headlines:
            print("No headlines found using 'gs-c-promo-heading'.")
            # Fallback approach: search for headings in 'h3', 'span', or any other container
            headlines = soup.find_all('h3')

        print("\nTop News Headlines:")
        for i, headline in enumerate(headlines[:5], 1):  # Limiting to the first 5 headlines
            print(f"{i}. {headline.get_text().strip()}")
        return headlines[:5]
    else:
        print("Failed to retrieve news headlines.")
        return None

# Main function to demonstrate both functionalities
def main():
    # Part 2: Web scraping to get news headlines
    scrape_news_headlines()

# Ensure the main function is executed when the script is run
if __name__ == "__main__":
    main()
```
