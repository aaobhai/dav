```python
# Define a function to simulate loading data (instead of reading from a file)
def load_data():
    # Sample data for testing
    data = [
        "hello world\n",
        "python is great\n",
        "this is a test\n",
        "goodbye world\n"
    ]
    print("Data loaded successfully.")
    return data

# Define a function to process the loaded data
def process_data(data):
    processed_data = []
    for line in data:
        processed_line = line.strip().upper()  # Convert each line to uppercase
        processed_data.append(processed_line)
    return processed_data

# Define a function to save the processed data (simulated by printing to console)
def save_data(data):
    print("Processed data:")
    for line in data:
        print(line)

# Main function to demonstrate the process
def main():
    # Load data (from the embedded sample data)
    data = load_data()

    # Process data
    if data:
        processed_data = process_data(data)

        # Save processed data (simulating file save by printing)
        save_data(processed_data)

# Ensure the main function is executed when the script is run
if __name__ == "__main__":
    main()
```
